```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
# Step 1: Training Data (Hours Studied, Marks Scored)
# We'll only use Hours Studied for simplicity
X = np.array([[2], [4], [6], [1]])   # Hours Studied
y = np.array(['Pass', 'Pass', 'Pass', 'Fail'])  # Result
```

```python
print(X)
print(y)
```

Step 1Ô∏è‚É£ ‚Äî Store the Training Data

KNN doesn‚Äôt learn a model in advance.

It just stores all the training data.

```python
# Step 2: Create KNN Model with K = 3
model = KNeighborsClassifier(n_neighbors=3)
```

Step 2Ô∏è‚É£ ‚Äî Choose the value of K (number of neighbors)

You decide how many neighbors you want the algorithm to check.

Example: K = 3 means we‚Äôll look at the 3 closest neighbors.

üìå Tip:

Small K (e.g., 1) can make the model sensitive to noise.

Bigger K smooths predictions.

```python
# Step 3: Train the Model (KNN just stores the data)
model.fit(X, y)
```

Step 3Ô∏è‚É£ ‚Äî Get a New Data Point (You Want to Predict)

Example:

A student studied 3 hours, will they Pass or Fail?

We don‚Äôt know ‚Äî we want KNN to predict.

Step 4Ô∏è‚É£ ‚Äî Calculate the Distance between New Data & All Stored Data
üî∏ Usually, we use Euclidean Distance:

Distance = sqrt_root(x_2 - x_1)^2
 
‚Äã
 
For each stored data point:

(3 hours) vs (2 hours) = Distance = 1

(3 hours) vs (4 hours) = Distance = 1

(3 hours) vs (6 hours) = Distance = 3

(3 hours) vs (1 hour) = Distance = 2

Step 5Ô∏è‚É£ ‚Äî Select the K Nearest Neighbors

If K = 3, we pick the 3 closest neighbors:

(2 hours) ‚Üí Pass

(4 hours) ‚Üí Pass

(1 hour) ‚Üí Fail

Step 6Ô∏è‚É£ ‚Äî Do Voting (for Classification)

Out of 3 neighbors: 2 say Pass, 1 says Fail

Majority wins!

Prediction = Pass

```python
# Step 4: Predict for a new student who studied 3 hours
new_data = [[3]]
prediction = model.predict(new_data)
print("üìå Predicted Result for 3 hours study:", prediction[0])
```

```python
# Step 5 (Optional): Visualize the Data Points and New Prediction
plt.figure(figsize=(8, 5))

# Plot Pass as green, Fail as red
# X = np.array([[2], [4], [6], [1]])
for i in range(len(X)):  #range(4) --> 0,1,2,3   -- i = 0
    # y = np.array(['Pass', 'Pass', 'Pass', 'Fail']) 
    if y[i] == 'Pass':  # y[0] = Pass
        plt.scatter(X[i], 0, color='green', label='Pass' if i == 0 else "")   #X[0] = 2
    else:
        plt.scatter(X[i], 0, color='red', label='Fail' if i == 3 else "")

# Plot the new data point
plt.scatter(new_data, 0, color='blue', marker='x', s=100, label='New Data (3 hours)')

plt.xlabel('Hours Studied')
plt.yticks([])
plt.title('KNN - Pass or Fail Prediction')
plt.legend()
plt.grid(True)
plt.show()
```

| Step | Action                                  |
| ---- | --------------------------------------- |
| 1    | Store the training data                 |
| 2    | Choose K                                |
| 3    | Get new data to predict                 |
| 4    | Calculate distance with all data points |
| 5    | Select K closest neighbors              |
| 6    | Do majority voting (classification)     |
| 7    | Make prediction                         |
