```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Input data
hours = np.array([[1], [2], [3], [4], [5]])   #2D array for hours
marks = np.array([35, 40, 45, 50, 55])   #1D array for marks
print(hours)
print(marks)
```

```python
print(hours.shape)
print(marks.shape)
```

```python
# Step 1: Create the model
model = LinearRegression()
```

```python
# Step 2: Train the model
model.fit(hours, marks)
```

```python
a = np.array([[6]])
print(a.shape)
```

```python
list_1 = [1,2]
list_1[0]
```

```python
# Step 3: Predict for 6 hours
predicted_6 = model.predict([[6]])
print("ðŸ“Œ Predicted marks for 6 hours of study:", predicted_6[0])
```

```python
# Step 3: Predict for 6 hours
predicted_7 = model.predict([[7]])
print("ðŸ“Œ Predicted marks for 6 hours of study:", predicted_7[0])
```

```python
# Step 4: Predict on training data (for error & plot)
predicted_train = model.predict(hours)    #1,2,3,4,5
```

```python
predicted_train
```

```python
# Step 6: Plot the data and regression line
plt.figure(figsize=(8, 5))
plt.scatter(hours, marks, color='blue', label='Actual Marks')
plt.plot(hours, predicted_train, color='red', label='Regression Line')
plt.scatter(6, predicted_6, color='green', label='Prediction for 6 hrs', marker='o', s=100)
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.title("Linear Regression - Hours vs Marks")
plt.legend()
plt.grid(True)
plt.show()
```
