```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Prepare dataset
data = {
    'Am_I_Hungry': ['Yes', 'Yes', 'Yes', 'No', 'No'],
    'Is_It_Weekend': ['Yes', 'No', 'Yes', 'Yes', 'No'],
    'Shall_I_Eat_Pizza': ['Yes', 'Yes', 'Yes', 'No', 'No']
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Convert Yes/No to 1/0
df_encoded = df.replace({'Yes': 1, 'No': 0})
```

```python
df_encoded.head()
```

```python
# Features and target
X = df_encoded[['Am_I_Hungry', 'Is_It_Weekend']]
y = df_encoded['Shall_I_Eat_Pizza']
```

```python
# Train Decision Tree
model = DecisionTreeClassifier(criterion='entropy', random_state=42, max_depth=3)
model.fit(X, y)
```

```python
# Plot the tree
plt.figure(figsize=(8,6))
plot_tree(model, feature_names=X.columns, class_names=['No', 'Yes'], filled=True)
plt.show()
```

```python
# Example Prediction
print("Prediction for Hungry=No, Weekend=Yes:", model.predict([[0, 1]]))
print("Prediction for Hungry=Yes, Weekend=No:", model.predict([[1, 0]]))
```
